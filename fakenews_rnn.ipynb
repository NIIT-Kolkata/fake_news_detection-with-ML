```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
```

```python
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3 and token not in stop_words:
            result.append(token)
            
    return result
```

```python
df['clean'] = df['text'].apply(preprocess)
```

```python
df
```

```python
list_of_words = []
for i in df['clean']:
    for j in i:
        list_of_words.append(j)
```

```python
total_words = len(list(set(list_of_words)))
total_words
```

```python
df['clean_joined'] = df['clean'].apply(lambda x: " ".join(x))
```

```python
df
```

```python
maxlen = -1
for doc in df.clean_joined:
    tokens = nltk.word_tokenize(doc)
    if(maxlen<len(tokens)):
        maxlen = len(tokens)
print( maxlen)
```

```python
df = pd.get_dummies(df, columns=['label'], drop_first=True)
```

```python
df
```

```python
x=df['clean_joined']
y=df['label_REAL']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.1,random_state=2021) 
```

```python
from nltk import word_tokenize
```

```python
tokenizer = Tokenizer(num_words = total_words)
tokenizer.fit_on_texts(x_train)
```

```python
train_sequences = tokenizer.texts_to_sequences(x_train)
test_sequences = tokenizer.texts_to_sequences(x_test)
```

```python
train_sequences[2]
```

```python
padded_train = pad_sequences(train_sequences,maxlen = 40, padding = 'post', truncating = 'post')
padded_test = pad_sequences(test_sequences,maxlen = 40, truncating = 'post')
```

```python
rnn_model = Sequential()
```

```python
rnn_model.add(Embedding(total_words, output_dim = 128))
```

```python
rnn_model.add(Bidirectional(LSTM(128)))
```

```python
rnn_model.add(Dense(128, activation = 'relu'))
rnn_model.add(Dense(1,activation= 'sigmoid'))
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
rnn_model.summary()
```

```python
rnn_model.fit(padded_train, y_train, batch_size =512,epochs =3)
```
